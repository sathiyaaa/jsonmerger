Json merger:


The files are in specific folder location in my device

There are 8 files namely i1,i2,i3,i4,i5,i6,i7,i8 with prefix 'i'
The output files are marked with prefix '0'

The total size of the file after combining individual files have been found out to be 748 bytes.

The maximum size is given as 300 bytes
The are three output files obtained of lengths 265,287,224 bytes long. (less than 300 bytes and has valid json objects)


```python
import json
import glob   #for file accessing
import os


#function for creating adding individual json files and splitting them 
def jsonmerge(folder,inputpath,outputpath,maxsize):
    oprefix="o"       #prefix of json output files
    osuffix=1             #suffix of json output files
    temp=[]               #for storing intermediary json objects
    jsontemp={}           #for storing intermediary json dict
    keydict = {}
    #collectivejson file creation
    print('File names and size are')
    for f in glob.glob(folder+inputpath):
        
        print(f)
        print(os.stat(f).st_size)
        with open(f, "r") as inf:
            infile=json.load(inf)
            for k in infile.keys():
                
                if not keydict:
                    keydict[k]=infile[k]
                else:
                    keydict[k]+=infile[k]

        #collective json file writing
        with open(folder+outputpath, "w") as outfile:
            json.dump(keydict, outfile)

    #combined file size
    print("Combined file size is")
    print(str(os.stat(folder+outputpath).st_size) + "bytes")
    
    #accessing collective json file for splitting
    with open(folder+outputpath,'r') as splitfile:
        splitted=json.load(splitfile)
    totalsize=0
    i=0
    for k,v in splitted.items():
        while(i<len(v)):
            inisize=6+len(k)   #6 the default charecters in the json object excluding key length 
            totalsize=inisize
            obsize=v[i]
            while((totalsize+len(json.dumps(obsize))+1)<maxsize):
                #print(obsize)
                temp.append(v[i])
                totalsize+=len(json.dumps(obsize))+1
                i+=1
                if(i<len(v)):
                    obsize=v[i]
                else:
                    break
                #print(totalsize)
            jsontemp[k]=temp
            #print(temp)
            #individual json file writing after split to sizes<maxsize
            
            print('Output file names are')
            with open(oprefix+str(osuffix)+".json",'w') as splitfile:
                json.dump(jsontemp,splitfile)
                print(oprefix+str(osuffix)+".json")
                osuffix+=1
            temp.clear()


folder="C:\\Users\\Sathiya Prakash\\"    
inputpath=input()                               
outputpath=input()
maxsize=300
jsonmerge(folder,inputpath,outputpath,maxsize)          
```
